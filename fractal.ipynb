```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('/content/drive/MyDrive/seoul_dental_hospital_xy.csv')
```

# 진행 순서
1. 원점 맞추기
2. 박스 그리기
3. 박스 개수 늘려가기
4. x^ = x - x_min
   y^ = y - y_min
5. log(박스개수), 채워진 박스개수 그래프 각각 그리기

```python
# 1. 시작점을 (0,0)으로 만들어주기 위해 각각 빼주었다
x = d['x'] - 180000
y = d['y'] - 435000
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.axis([0, 40000, 0, 40000])
plt.axhline(20000, color='r') # y축 그리기
plt.axvline(20000, color='r') # x축 그리기
plt.show()
```
